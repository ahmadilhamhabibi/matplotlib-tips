# Adding Text

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 1, 4, 6, 4])
```

## Add Text

### Basic Text

```python
plt.scatter(x, y)
plt.text(4, 5.5, "Peak Point");
```

```python
plt.scatter(x, y)
plt.text(4, 5.5, "Peak Point")
plt.text(2, 1.3, "Low Point");
```

## Horizontal and Vertical Alignment: ``ha``, ``va``

```python
plt.scatter(x, y)
plt.text(4, 5.5, "Peak Point", ha="center")
plt.text(2.2, 1, "Low Point", va="center");
```

## Styling

### Color and Font Properties

```python
students = ['A', 'B', 'C', "D", "E"]
grades = [80, 42, 68, 91, 77]

plt.bar(students, grades, edgecolor='black', color='xkcd:light gray blue')
plt.axhline(100, color='xkcd:brick red', lw=3)

plt.text(2.5, 104, "Maximum Grade",
         fontsize=14,
         color='xkcd:brick red',
         fontweight='bold',
         fontstyle='italic'
        )
    
plt.xlabel('Student')
sns.despine();
```

### Bounding Box: ``bbox``

```python
from matplotlib.patches import Circle

line_x = np.linspace(0, 10, 1001)
line_y = 5 - 2*np.abs(line_x - 5)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(line_x, line_y, lw=3)
plt.gca().margins(0.1)

plt.text(6, 5, "Critical Point",
        bbox={'facecolor': 'red', 'alpha': 0.3}
        )

cir = Circle((5, 5), 0.5, color='red', lw=3, fill=False, zorder=10)
plt.gca().add_patch(cir);
```

## Text in General Position or at Variable

### General Position: ``transform=ax.transAxes``

```python
def add_mean_line(data):
    mu = np.mean(data)
    
    ax=plt.gca()
    ax.axvline(mu, color='black', lw=3, ls='--')
    
    ax.text(0.7, 0.9, f'mean={mu:.2f}',
            transform=ax.transAxes,   #transforms positions to range from (0,0) to (1,1)
            fontweight='bold', fontsize=12)
```

```python
values = np.random.normal(10, 2, size=100)

plt.hist(values, color='lightblue')
add_mean_line(values);
```

### Variable in a Loop

```python
students = ['A', 'B', 'C', 'D', 'E']
grades = [80, 42, 68, 91, 77]

plt.bar(students, grades, edgecolor='black', color='xkcd:light gray blue')

for i, g in enumerate(grades):
    plt.text(i, g-8, f'{g}', ha='center', fontsize=16)
    
plt.xlabel('Student')
plt.yticks([])
sns.despine(left=True)
```
