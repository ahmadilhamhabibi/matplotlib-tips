# Changing the Figure Size

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(0, 4*np.pi, 1000)
Y = np.sin(X)
```

```python
plt.plot(X, Y)
plt.show()
```

## figsize

Reference the figure and its ``figsize`` argument to update the figure size of an individual figure.

The tuple you pass will be read in as (width in inches, width in height).

```python
plt.figure(figsize=(12,8))
plt.plot(X, Y)
plt.show()
```

```python
plt.figure(figsize=(8,12))
plt.plot(X, Y)
plt.show()
```

## via rcParams

Update your run command (rc) parameters to change the size of ALL future figures.

```python
plt.plot(X, Y)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (4, 4)
```

```python
plt.plot(X, Y)
plt.show()
```

```python
plt.plot(X, -Y)
plt.show()
```
