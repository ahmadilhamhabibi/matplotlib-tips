# Saving a Figure

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(99)

X = np.random.rand(100)
Y = 2*X + 0.5*np.random.rand(100)
```

```python
plt.scatter(X, Y)
plt.show()
```

## savefig

```python
plt.scatter(X, Y)
plt.savefig("random_scatter.png")
plt.show()
```

Matplotlib saves to **current directory** by default. But you can change the folder.

```python
%ls
```

Specify **file path** for other directory

```python
plt.scatter(X, Y)
plt.savefig("images/random_scatter.png")
plt.show()
```

## File Options - png, pdf, svg, ps

```python
plt.scatter(X, Y)
plt.savefig("images/random_scatter.svg")
plt.show()
```

## Increase Resolution

For higher resolution graphics, increase the **dots per inch (dpi)**. Default dpi is 100.

```python
plt.scatter(X, Y)
plt.savefig("images/random_scatter_highres.svg", dpi=200)
plt.show()
```

## Figure Cutting Off

```python
plt.scatter(X, Y)
plt.xlabel("Random Data", fontsize=24)
plt.savefig("images/random_scatter_xlabel_fixed.svg", bbox_inches="tight")
plt.show()
```
