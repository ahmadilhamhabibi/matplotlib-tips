# Add Hatching

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
values = [20, 40, 60]
```

## Hatching on Barplots

```python
plt.bar(range(len(values)), values, hatch="/");
```

### Patterns

```python
plt.bar(range(len(values)), values, hatch="x");
```

```python
plt.bar(range(len(values)), values, hatch="*");
```

### Patterns List

```python
patterns = ['', '\\', '']
bars = plt.bar(range(len(values)), values)

for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern);
```

```python
type(bars)
```

## hatching on Other Objects

### Patches: Rectangle, Circle

```python
from matplotlib.patches import Rectangle, Circle
```

```python
plt.figure(figsize=(5, 5))
plt.xlim(0, 10)
plt.ylim(0, 10)

rect = Rectangle((5, 5), 4, 4, fill=False, color='orange', lw=4, hatch='o')

cir = Circle((3, 3), 2, fill=False, color='teal', lw=4, hatch='-')

plt.gca().add_patch(rect)
plt.gca().add_patch(cir);
```

### Pie Wedge

```python
wedges, labels = plt.pie(values)

wedges[0].set_hatch('|');
```

```python
type(wedges[0])
```

## ``fill_between``

```python
x = np.linspace(0, 10, 100)
y1 = 25 - (5-x)**2
y2 = -y1
```

```python
plt.fill_between(x, y1, y2, hatch='.');
```

## Styling

### Color

```python
plt.bar(range(len(values)), values, hatch='/', 
        facecolor='lightgray',
        edgecolor='deeppink'
       );
```

```python
plt.rcParams['hatch.color'] = 'blue'
```

```python
plt.bar(range(len(values)), values, hatch='/', 
        facecolor='lightgray'
       );
```

### Line Width

```python
plt.rcParams['hatch.linewidth'] = 6
```

```python
plt.bar(range(len(values)), values, hatch='/', facecolor='lightgray');
```
